```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
iterations = []
train_acc = []
train_loss = []
val_acc = []
val_loss = []

with open(r'result_data\ptbdb0409\42\self-onn_Fcanet_ada16_64_19_mappar\2025-04-10-09-29-07\plot\training_results.txt','r') as f:
    for line in f:
        # 解析每行数据
        parts = line.strip().split(', ')
        metrics = {}
        for part in parts:
            key, value = part.split(': ')
            metrics[key.strip()] = value.strip()
        
        # 存储解析结果
        iterations.append(int(metrics['Iteration']))
        train_acc.append(float(metrics['Train Accuracy']))
        train_loss.append(float(metrics['Train Loss']))
        val_acc.append(float(metrics['Validation Accuracy']))
        val_loss.append(float(metrics['Validation Loss']))
```

```python
plt.figure(figsize=(12, 6))

# 准确率曲线子图
plt.subplot(1, 2, 1)
plt.plot(iterations, train_acc, 'b-', linewidth=2, label='Train Acc')
plt.plot(iterations, val_acc, 'r--', linewidth=2, label='Val Acc')
plt.xlabel('Iterations')
plt.ylabel('Accuracy (%)')
plt.title('Training vs Validation Accuracy')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.ylim(0, 100)  # 固定y轴范围

# 损失曲线子图
plt.subplot(1, 2, 2)
plt.plot(iterations, train_loss, 'g-', linewidth=2, label='Train Loss')
plt.plot(iterations, val_loss, 'm--', linewidth=2, label='Val Loss')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# 调整布局
plt.tight_layout()
plt.show()
```

```python
data = []
with open(r'result_data\ptbdb0409\42\self-onn_Fcanet_ada16_64_19_mappar\2025-04-10-09-29-07\plot\training_results.txt','r') as file:
    for line in file:
        parts = line.strip().split(', ')
        iteration = int(parts[0].split(': ')[1])
        train_accuracy = float(parts[1].split(': ')[1])
        train_loss = float(parts[2].split(': ')[1])
        val_accuracy = float(parts[3].split(': ')[1])
        val_loss = float(parts[4].split(': ')[1])
        data.append({
            'Iteration': iteration,
            'Train Accuracy': train_accuracy,
            'Train Loss': train_loss,
            'Validation Accuracy': val_accuracy,
            'Validation Loss': val_loss
        })
```
